# Notebook for a prediction ML model
- This model use a `melb_data.csv` like a dataset
- The data set the melbourne houses price

```python
import pandas as pd

melb_df = pd.read_csv('melb_data.csv')
melb_df
```

```python
melb_df.describe()
```

## Limpieza de datos


```python
melb_df = melb_df.dropna(axis = 0)
melb_df
```

## Machine learning


```python
y = melb_df['Price']
x = melb_df[['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']]
```

### Arbol de decisiones

```python
from sklearn.tree import DecisionTreeRegressor
from datetime import datetime

melbourne_model = DecisionTreeRegressor()
melbourne_model.fit(x, y)
fecha_entrenamiento = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
```

### Regresion linel

```python
from sklearn.linear_model import LinearRegression


melbourne_lineal = LinearRegression(n_jobs=-1)
melbourne_lineal.fit(x, y)

print("Predicciones para las 5 primeras casas (Regresión Lineal):")
print(x.head())

print("El precio real es:")
print(y.head())

print("Las predicciones son:")
print(melbourne_lineal.predict(x.head()))
```

```python
print("Predicciones para las 5 primereas casas: ")
print(x.head(100))
print("El precio real es: ")
print(y.head(100))
print("Las predicciones son")
print(melbourne_model.predict(x.head(100)))
```

## MAE
Ver las metricas de errores absolutos

```python
from sklearn.metrics import mean_absolute_error

predicted_home_prices = melbourne_model.predict(x)
mean_absolute_error(y, predicted_home_prices)
```

## Validacion

```python
from sklearn.model_selection import train_test_split

train_X, val_X, train_Y, val_y = train_test_split(x, y)

melbourne_model = DecisionTreeRegressor()
melbourne_model.fit(train_X, train_Y)

val_predictions = melbourne_model.predict(val_X)
print(mean_absolute_error(val_y, val_predictions))
```

```python
from sklearn.ensemble import RandomForestRegressor

forest_model = RandomForestRegressor()
forest_model.fit(train_X, train_Y)
melb_preds = forest_model.predict(val_X)
print(mean_absolute_error(val_y, melb_preds))
```

# Manejo de SQL y guardado de la data de errores y mejora de entrenamientos

### Creacion de la base de datos

```python
import sqlite3 as sql

con = sql.connect("newmodel.db")
cur = con.cursor()

# Creacion de la tabla principal
cur.execute('''CREATE TABLE IF NOT EXISTS model_error (
            id INTEGER PRIMARY KEY,
            entrenamiento INT,
            error INT
            )
''')

con.commit()
```

**Agregar los datos a la base de datos progresivamente**

```python
from datetime import datetime
import sqlite3 as sql

# Fecha actual
fecha_entrenamiento = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
err = mean_absolute_error(val_y, melb_preds)

# Usar context manager para asegurar que la conexión se cierre
with sql.connect("newmodel.db", timeout=30.0) as con:
    cur = con.cursor()
    cur.execute('''
    INSERT INTO model_error (entrenamiento, error)
    VALUES (?, ?)
    ''', (fecha_entrenamiento, err))
    con.commit()
```

**Sacar los datos de la tabla de modelia**


```python
df_sql = pd.read_sql_query(
    "SELECT * FROM model_error",
    con
)

df_sql["entrenamiento"] = pd.to_datetime(df_sql["entrenamiento"])

# Ordenar por fecha (importante)
df_sql = df_sql.sort_values("entrenamiento")
print(df_sql)
```

```python
con.commit()
cur.close()
con.close()
```

## Graficos de errores y avance

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(10,5))
plt.plot(df_sql["entrenamiento"], df_sql["error"])

plt.title("Evolución del error del modelo")
plt.xlabel("Fecha de entrenamiento")
plt.ylabel("Mean Absolute Error (MAE)")
plt.xticks(rotation=45)

plt.grid(True)
plt.tight_layout()
plt.show()
```
