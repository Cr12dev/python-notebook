```python
import pandas as pd

melb_df = pd.read_csv('melb_data.csv')
melb_df
```

```python
melb_df.describe()
```

## Limpieza de datos


```python
melb_df = melb_df.dropna(axis = 0)
melb_df
```

## Machine learning


```python
y = melb_df['Price']
x = melb_df[['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']]
```

### Arbol de decisiones

```python
from sklearn.tree import DecisionTreeRegressor

melbourne_model = DecisionTreeRegressor()
melbourne_model.fit(x, y)
```

### Regresion linel

```python
from sklearn.linear_model import LinearRegression


melbourne_lineal = LinearRegression(n_jobs=-1)
melbourne_lineal.fit(x, y)

print("Predicciones para las 5 primeras casas (Regresión Lineal):")
print(x.head())

print("El precio real es:")
print(y.head())

print("Las predicciones son:")
print(melbourne_lineal.predict(x.head()))
```

```python
print("Predicciones para las 5 primereas casas: ")
print(x.head())
print("El precio real es: ")
print(y.head())
print("Las predicciones son")
print(melbourne_model.predict(x.head(100)))
```

## MAE
Ver las metricas de errores absolutos

```python
from sklearn.metrics import mean_absolute_error

predicted_home_prices = melbourne_model.predict(x)
mean_absolute_error(y, predicted_home_prices)
```

## Validacion

```python
from sklearn.model_selection import train_test_split

train_X, val_X, train_Y, val_y = train_test_split(x, y)

melbourne_model = DecisionTreeRegressor()
melbourne_model.fit(train_X, train_Y)

val_predictions = melbourne_model.predict(val_X)
print(mean_absolute_error(val_y, val_predictions))
```

```python
from sklearn.ensemble import RandomForestRegressor

forest_model = RandomForestRegressor()
forest_model.fit(train_X, train_Y)
melb_preds = forest_model.predict(val_X)
print(mean_absolute_error(val_y, melb_preds))
```

## Graficos 


### Por regresion de arbol

```python
import matplotlib.pyplot as plt
import numpy as np

# Tomar las primeras 100 casas
x_100 = x.head(100)
y_100 = y.head(100)

# Predicciones
predictions_100 = melbourne_model.predict(x_100)

# Índices de las casas
houses = np.arange(1, 101)

# Crear gráfica
plt.figure(figsize=(15, 6))

# Precios reales
plt.plot(houses, y_100.values, label='Precio real')

# Predicciones
plt.plot(houses, predictions_100, label='Predicción')

# Títulos y etiquetas
plt.title("Precios reales vs Predicciones (100 primeras casas)")
plt.xlabel("Número de casa")
plt.ylabel("Precio")

# Leyenda
plt.legend()

# Cuadrícula
plt.grid(True)

# Mostrar gráfica
plt.show()
```
