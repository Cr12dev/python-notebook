## Mini algoritmo de laboratorio

Comprueba la empresa y si es una la categoriza como un estado o como otro


```python
import pandas as pd

data = {
    "business": ["Zyrova", "Github", "Netflix", "Amazon"],
    "Revenues": [3e3, 50e3, 100e3, 100e3 * 10],
    "emails": ["zyrova@support.com", "support@github.com", "netflix@support.com", "amazon@support.com"],
}

df = pd.DataFrame(data)

def algoritmo(row):
    if row["business"] == 'Zyrova':
        return "Especial"
    else:
        return "Normal"

# Crear nueva columna
df["estatus"] = df.apply(algoritmo, axis=1)

print(df)
```
