# Base de datos PL/SQL en python
Libreta para aprender como funcionan las bases de datos **PL/SQL** en python con la libreria sqlite3


## Uso basico de SQL


```python
import sqlite3

# Crear la conexion a la base de datos, si el fichero .db no existe se creará uno nuevo
con = sqlite3.connect("databse.db")
cur = con.cursor()

# Crear una tabla en la base de datos
cur.execute('''CREATE TABLE usuarios (
    id INTEGER PRIMARY KEY,
    nombre TEXT,
    puntos REAL 
)''')

```

### Sacar los nombre de la tabla usuarios


```python
# Llamar a la tabla de la base de datos
cur.execute('''SELECT nombre FROM usuarios''')
for row in cur.fetchall():
    print(row[0])
```

### Insertar nombres en la tabla usuarios


```python
# Insertar un nuevo dato, en este caso un nombre en la base de datos
cur.execute('''INSERT INTO usuarios (nombre)
            VALUES ('Thomas');''')
```

### Eliminar un nombre en especifico en la base de datos


```python
cur.execute('''DELETE FROM usuarios WHERE nombre = 'Jhon';
''')
```

### Actualizar nombre o datos en la tabla de usuarios

```python
cur.execute('''UPDATE usuarios SET nombre = 'NuevoJhon' WHERE nombre = 'Thomas';
''')
```
