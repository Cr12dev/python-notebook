### Problema 1:
En el parque se ha observado que los dinosaurios carnívoros depositan un número par de huevos
. Debemos crear una función que reciba una lista de cantidades de huevos y sume únicamente aquellos que pertenecen a carnívoros (los pares)
.

```python
def sumar_huevos_carnivoros(huevos: list[int]) -> int:
    total = 0

    # Bucle para ver como es cada huevo
    for i, cantidad in enumerate(huevos):
        # Verificar la cantidad par del huevo
        if cantidad % 2 == 0:
            print(f"Dinosaurio en indice {i} es carnivoro con {cantidad} huevos.")
            total += cantidad
    
    return total

#Datos para la prueba
huevos = [14]
result = sumar_huevos_carnivoros(huevos)
print(f"La suma total de huevos carnivoros es: {result}")
```
